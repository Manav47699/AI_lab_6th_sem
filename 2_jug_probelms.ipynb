```python
from collections import deque
```

```python
jug_a = 4
jug_b = 3
```

```python
def get_next_state(state):
    a, b = state         #seperating the tuple (0, 0)
    next_state = []

    next_state.append((jug_a, b))      #fill a

    next_state.append((a, jug_b))      #fill b

    next_state.append((0, b))       #empty a
     
    next_state.append((a, 0))       #empty b

    transfer = min (a, jug_b -b)                   #min(water available in jug_a, empty space in jug_b) calculating how much can be poured from a to b
    next_state.append((a-transfer, b+transfer))     #pour a -> b
 
    transfer = min(b, jug_a - a)                     
    next_state.append((a + transfer, b- transfer))     #pour b -> a

    return next_state
    
```

```python
#breadth first search functon
def bfs():
    start = (0,0)

    queue = deque([(start, [start])])       #stores (current state, path taken)
    visited = {}

    while queue:                            #run untill queue is empty
        state, path = queue.popleft()

        if state in visited:
            continue

        visited.add(state)

        if state[0] == 2:              #checking if the jug_a has 2 liters of water, (final state)
            return path

        for next_state in get_next_state(state):
            if next_state not in visited:
                queue.append((next_state, path + [next_state]))

    return None
```

```python
solution = bfs()

print ("Solution is :")

for state in solution:
    print (state)
```
