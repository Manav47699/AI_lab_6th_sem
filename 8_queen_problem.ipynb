```python
from pprint import pprint
n = 4
board = [['_']*n for i in range (n)]
pprint (board)
```

```python
#this function return if it is safe fot the queen to placed in the specific place
def is_safe(board, row, col, n):
    #column test
    for x in range (row):
        if board [x][col] == 'Q':
            return False
        # top left diagonal test
        r = row -1 
        c = col -1
        while (r >= 0 and c >= 0):
            if board[r][c] == 'Q':
                return False
            r= r-1
            c = c-1
        #top right diagonal test
        r = row -1 
        c = col +1
        while (r >= 0 and c< n):
            if board[r][c] == 'Q':
                return False
            r= r-1
            c = c+1
    return True




```

```python
print (is_safe(board, 1, 2, n))
```
