```python
from collections import deque
```

```python
def is_valid(state):
    F, W, G, C = state          #spliting the tubples

    if W == G and F!= W:           #checking if wolf and goat are together without the farmer. If this condition is not fulfilled , the state can't progress
        return False
    
    if G == C and F != G:
        return False
    
    return True
```

1-f changes the sides

- For example, if the farmer is east i.e f = 0, then 1-0 = 1 (west). So this is east -> west

- if the farmer is west i.e f =1 , then 1 - 1 = 0 (east). So this is west -> east


```python
def get_next_state(state):
    F, W, G, C = state
    next_state = []

# Farmer crosses alone
    new_state = (1-F, W, G, C)
    if is_valid(new_state):
        next_state.append(new_state)


#farmer crosses with wolf
    if F == W:
        new_state = (1-F, 1-W, G, C)          
        if is_valid(new_state):
            next_state.append(new_state)


#farmer crosses with goat
    if F == G:
        new_state = (1-F, W, 1-G, C)
        if is_valid(new_state):
            next_state.append(new_state)



#farmer crosses with cabbage
    if F == C:
        new_state = (1-F, W, G, 1-C)
        if is_valid(new_state):
            next_state.append(new_state)

    return next_state

```

```python

# 0 -> east and 1 -> west
def bfs():
    start = (0,0,0,0)
    goal = (1,1,1,1)

    queue = deque([(start,[start])])      #(current_state, path)
    visited = {}

    while queue:
        state, path = queue.popleft()

        if state in visited:
            continue

        visited.add(state)

        if state == goal:
            return path

        for next_state in get_next_state(state):
            if next_state not in visited:
                queue.append((next_state, path+[next_state]))

    return None
```

```python
solution = bfs()

print("Solution Path:")
for state in solution:
    print(state)
```
